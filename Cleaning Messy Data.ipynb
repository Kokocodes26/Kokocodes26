# Handling missing values

```python
#Drop the missing values (THIS IS A NO NO) based on the situation
employee_data.dropna() #.dropna() only drops it on the applicable dataset. Dropna(inplace=true) will permanently drop from the dataset
```

```python
#Filling in the missing values
#cleaning up the age column
employee_data['age'] = employee_data.age.fillna(employee_data.age.mean())
```

```python
employee_data
```

```python
#cleaning up the income column
employee_data['income'] = employee_data.income.fillna(employee_data.income.median())
```

```python
employee_data
```

```python
#cleaning up the gender column
employee_data['gender'] = employee_data['gender'].fillna(employee_data['gender'].mode()[0])
```
