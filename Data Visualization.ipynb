Scatter Plot

```python
#Step1: creating a fake dataset
x = [1, 3, 5, 7, 9, 11]
y = [10, 25, 36, 35, 48, 61]

#Step 2: plotting the dataset
plt.scatter(x,y,c='green', marker='*')

#Step 3: add the x and y labels
plt.xlabel('this is the x-axis')
plt.ylabel('this is the y-axis')

#Step 4: add the graph title
plt.title('Scatter Plot Example')

#Step 5: show the graph
plt.show()
```

Pie Plot

```python
#Step 1: Creating a fake dataset
subjects = ['Python for DA', 'Data Modeling', 'MS T-SQL', 'Excel for DA']
scores = [95, 85, 98, 50]

#Step 2: plot the dataset
plt.pie(scores, labels=subjects, colors=['red', 'blue', 'green', 'yellow'], shadow=True, explode=(0,0.1,0,0.2))

#Step 3: Add the graph title
plt.title('This is the title for Pie Graph')

#Step 4: Add show the graph
plt.show()
```

```python
#Step 1: Creating a fake dataset
subjects = ['Python for DA', 'Data Modeling', 'MS T-SQL', 'Excel for DA']
scores = [95, 85, 98, 50]

#Step 2: plot the dataset
plt.pie(scores, 
        labels=subjects, 
        colors=['red', 'blue', 'green', 'yellow'], 
        shadow=True, 
        explode=(0,0.1,0,0.2),
        startangle=90,
        autopct='%1.1f%%')

#Step 3: Add the graph title
plt.title('This is the title for Pie Graph')

#Step 4: Add show the graph
plt.show()
```

Bar plot

```python
#Step 1: Creating a fake dataset
movie_rating = [1, 2, 3, 4, 5]
rating_counts = [21, 45, 72, 89, 40]

#Step 2: plot the dataset
plt.bar(movie_rating, rating_counts, color='purple')

#Step 3: Add the x and y labels
plt.xlabel('Movie rating')
plt.ylabel('Movie rating counts')

#Step4: Add the graph title
plt.title('Distribution of Movie Ratings')

#Step 5: Add show the graph
plt.show()
```
