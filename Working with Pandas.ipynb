Join in Python

```python
#Step 1: importing the required library
import pandas as pd
import numpy as np
```

```python
#loading our dataset
dest_df = pd.read_csv('/Users/kosi_og/Python/Datasets/dest.csv')
tips_df = pd.read_csv('/Users/kosi_og/Python/Datasets/tips.csv')
```

```python
#Displaying the dataframe for dest
dest_df.head()
```

```python
#Displaying the dataframe for tips
tips_df.head()
```

```python
#Inner Join - returns matching records for two tables
inner_join_df = pd.merge(dest_df, tips_df, on='EmpNr', how='inner')
```

```python
inner_join_df
```

```python
#Left Join - returns all the records from the left, and matching records on the right
left_join_df = pd.merge(dest_df, tips_df, on='EmpNr', how='left')
```

```python
left_join_df
```

```python
#Right JOin - returns all the records from the right, and matching records on the left
right_join_df = pd.merge(dest_df, tips_df, on='EmpNr', how='right')
```

```python
right_join_df
```

Filtering dataframe by rows

```python
left_join_df[left_join_df['Dest']=='Amsterdam']
```

```python
right_join_df[right_join_df['Amount']==2.5]
```
